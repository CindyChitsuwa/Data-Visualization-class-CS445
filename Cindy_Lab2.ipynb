Dataset: Maternal Health Risk

This dataset is about predicting maternal health risks using medical measurements. It includes data collected from pregnant women such as their blood pressure, blood sugar levels, age, body temperature, and heart rate. The goal is to classify whether a woman falls under low, mid, or high maternal health risk.

```python
# Import the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import seaborn

import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("Maternal Health Risk Data Set.csv")

# Preview the data
data.head(11)
```

The  countplot below shows how the risk levels (Low, Mid, High) are distributed in the dataset.

The countplot gives an overview of how balanced or imbalanced the classes are. In this case, Low Risk appears the most frequent, followed by Mid Risk, and High Risk is the least.

```python
sns.set(style="whitegrid")  # set background style
plt.figure(figsize=(6, 4))  # set figure size

sns.countplot(x="RiskLevel", data=data, palette="Set2")  # create count plot for risk levels

plt.title("Distribution of Maternal Health Risk Levels")  # title
plt.xlabel("Risk Level")  # x-axis label
plt.ylabel("Count")  # y-axis label

plt.tight_layout()  # adjust spacing
plt.savefig("Cindy_RiskLevelDistribution.png")  # save plot
plt.show()  # display plot
```

Age vs Risk Level

This histplot visualizes how age is distributed across different maternal health risk levels.
The histogram shows that Low Risk cases are more common in younger women, while High Risk tends to occur more in older age groups. It helps confirm that age may be a factor contributing to the increased health risks during pregnancy.

```python
plt.figure(figsize=(6, 4))  # set figure size

# create histogram of Age by Risk Level
sns.histplot(data=data, x="Age", hue="RiskLevel", multiple="stack", palette="Set2", edgecolor=".3")

plt.title("Age Distribution by Risk Level")  # add title
plt.xlabel("Age")  # x-axis label
plt.ylabel("Count")  # y-axis label

plt.tight_layout()  # adjust spacing
plt.savefig("Cindy_AgeDistribution_by_RiskLevel.png")  # save plot
plt.show()  # display plot
```

Average Blood Sugar by Risk Level
The barplot below shows that the high risk group has the highest average blood sugar, while the low Risk group has the lowest.
This suggests that blood sugar might be a strong factor of maternal health risk. Since the difference is visually clear, it could help in early detection or risk assessment.

```python
plt.figure(figsize=(6, 4))  # set figure size

# create barplot showing average blood sugar per risk level
sns.barplot(x="RiskLevel", y="BS", data=data, palette="Set2", estimator="mean")

plt.title("Average Blood Sugar by Risk Level")  # add title
plt.xlabel("Risk Level")  # x-axis label
plt.ylabel("Average Blood Sugar")  # y-axis label

plt.tight_layout()  # adjust spacing
plt.savefig("Cindy_AvgBloodSugar_vs_RiskLevel.png")  # save plot
plt.show()  # display plot
```

Feature Correlation Map

This correlation heatmap helps spot strong relationships between features. For example, SystolicBP and DiastolicBP are positively correlated, which makes sense since both are blood pressure measurements. 

```python
plt.figure(figsize=(8, 6))  # set figure size

# create heatmap of feature correlations (excluding RiskLevel)
sns.heatmap(data.drop("RiskLevel", axis=1).corr(), annot=True, cmap="Blues", linewidths=0.5)

plt.title("Feature Correlation Heatmap")  # add title
plt.tight_layout()  # adjust spacing
plt.savefig("Cindy_CorrelationHeatmap.png")  # save plot
plt.show()  # display plot
```

Relationships Between Features Colored by Risk

The pairplot gives a holistic view of the dataset. Coloring by RiskLevel helps you spot clusters and possible separations. It looks like some groups overlap a lot (like Mid and Low), while others (like High Risk) are easier to separate based on combinations like Age and HeartRate.

```python
# create pairplot to show relationships between features, colored by risk level
sns.pairplot(data, hue="RiskLevel", palette="husl", plot_kws={"alpha": 0.6})

plt.savefig("Cindy_Pairplot_RiskLevel.png")  # save plot
plt.show()  # display plot
```
